```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("diffs.txt", "r") as f:
    info_str = f.read()
```

```python
time = [float(i.split()[0]) for i in info_str.split("\n") if len(i) > 0]
diff = [float(i.split()[1]) for i in info_str.split("\n") if len(i) > 0]
```

```python
plt.plot(time, diff)
plt.show()
```
